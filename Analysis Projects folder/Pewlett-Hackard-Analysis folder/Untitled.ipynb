```python
# Dependencies
import os
import pandas as pd
import dataframe_image as dfi  # to save DataFrames as image files
```

```python
# Define a function that will display the head and tail of a DataFrame with a row of ellipses in between
def ends(df, x=5, y=5):
    return pd.concat([
        df.head(x),
        # Insert a single-row DataFrame with the same columns as df and
        # all '…' values (including the index)
        pd.DataFrame(columns=df.columns, index=['…'], data=[['…' for i in df.columns]]),
        df.tail(y)
    ])
```

```python
# Import SQL output CSVs into DataFrames
df_retirement_titles = pd.read_csv(os.path.join('.','data','retirement_titles.csv'))
df_unique_titles = pd.read_csv(os.path.join('.','data','unique_titles.csv'))
df_retiring_titles = pd.read_csv(os.path.join('.','data','retiring_titles.csv'))
df_mentorship_eligibilty = pd.read_csv(os.path.join('.','data','mentorship_eligibilty.csv'))
df_mentorship_titles = pd.read_csv(os.path.join('.','data','mentorship_titles.csv'))
df_mentors_vs_retirees = pd.read_csv(os.path.join('.','data','mentors_vs_retirees.csv'))

# Make DataFrame index columns start at 1
df_retirement_titles.index += 1
df_unique_titles.index += 1
df_retiring_titles.index += 1
df_mentorship_eligibilty.index += 1
df_mentorship_titles.index += 1
df_mentors_vs_retirees.index += 1
```
